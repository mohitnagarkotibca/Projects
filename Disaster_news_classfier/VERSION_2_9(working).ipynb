## What role does length of tweet plays in Disaster tweet and Non-Disaster tweets ?

```python
#Comparing length of real tweet and non real tweetss
plt.style.use('fivethirtyeight')
train_group=train_set_orig.groupby('target')

real_text_length= [len(text) for text in train_group.get_group(1)['text']]
neg_text_length= [len(text) for text in train_group.get_group(0)['text']]

bins=120

plt.figure(figsize=(20,5))

plt.hist(neg_text_length,bins=bins,label='fake',color='red',alpha=0.6)
plt.hist(real_text_length,bins= bins,label='real',color='green')

plt.ylabel('Number of characters')
plt.xlabel('Tweet count')
plt.title('affect of tweets length on Genuinity ')
plt.legend()
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(10,5))
plt.title('Number of characters in real and fake tweets')
ax1.hist(real_text_length,color= 'green')
ax1.set_xlabel('Number of real tweets',color= 'green')
ax1.set_ylim([0,1000])
ax2.hist(neg_text_length,color='red')
ax2.set_xlabel('Number of fake tweets',color='red')
ax2.set_ylim([0,1000])
```

## Answer:
## People Tend to write tweets little long when they are talking about a Real Disaster
It can be clearly seen that tweets size between 120 to 400 length are same in both type of tweets.


```python
def avg(x):
    return np.sum([len(word) for word in x])

neg_tweet_avg=  [np.sum(list(map(avg,text)))/len(text.split()) for text in train_group.get_group(0)['text']]
post_tweet_avg= [np.sum(list(map(avg,text)))/len(text.split()) for text in train_group.get_group(1)['text']]
```

```python
fig,(ax1,ax2)= plt.subplots(1,2,figsize=(15,5))
sns.histplot(neg_tweet_avg,color='red',ax=ax1,kde=True)
ax1.set_title('fake tweets')
ax1.set_xlabel('average tweet length')
sns.histplot(post_tweet_avg,color='green',ax=ax2,kde=True)
ax2.set_title('real tweets')
ax2.set_xlabel('average tweet length')
```

```python
class NeuralNetwork_2layer_1H:
    def __init__(self,num_iters,lr):
        self.lr= lr
        
        self.W1= None
        self.DW1= None

        self.W2= None
        self.DW2= None
        
        self.B1= None
        self.DB1= None
        
        self.B2= None
        self.DB2= None
        
        self.DZ1= None
        self.DZ2= None
        
        self.num_iters= num_iters
        self.N= None
        
        self.X= None
        self.Y= None
        self.A2=None
        
    def initialize_parameters(self,X,Y,nodes):
        self.X= X
        self.Y= Y
        self.W1= np.random.randint(5,size= (nodes,X.shape[1])) *0.01
        self.B1= np.zeros((nodes,1))
        self.W2= np.random.randint(5,size=((1,nodes))) *0.01
        self.B2= np.zeros((1,1))
        self.N= X.shape[0]        

    
    def Print_parameters(self):
        print('Initializing parameters..')
        print('W1:',self.W1.shape,'DW1:',self.DW1.shape)
        
        print('B1:',self.B1.shape,'DB1:',self.DB1.shape)
        
        
        print('W2:',self.W2.shape,'DW2:',self.DW2.shape)
        
        print('B2:',self.B2.shape,'DB2',self.DB2.shape)
        
        print('Z1:',self.Z1.shape,'DZ1:',self.DZ1.shape)
        
        print('Z2:',self.Z2.shape,'DZ2:',self.DZ2.shape)
        print()
        
    def sigmoid(self,z):
        return 1/(1+np.exp(-z))
    
    def feed_forward(self):
        self.Z1= np.dot(self.W1,self.X.T)+self.B1

        A1= np.tanh(self.Z1)

        self.Z2= np.dot(self.W2,A1)+self.B2
        
        self.A2= self.sigmoid(self.Z2)
        cache= self.A2

        self.DZ2= self.A2 -self.Y
        
        self.DW2= (1/self.N)*np.dot(self.DZ2,A1.T)
        self.DB2= (1/self.N)*np.sum(self.DZ2,axis=1)
        
        d_act= np.multiply(self.Z1 , (1-self.Z1))
        self.DZ1= (1/self.N)*np.multiply(np.dot(self.W2.T,self.DZ2), d_act)
        
        self.DW1= (1/self.N)*np.dot(self.DZ1,self.X)
        self.DB1= (1/self.N)*np.sum(self.DZ1,axis=1)
        
        return cache
    
    def find_cost(self):
        a= np.ravel(np.dot(self.Y , np.log(self.A2).T))
        b= np.ravel(np.dot((1-self.Y),np.log(1-self.A2.T)))
        cost = -(1/self.N) * (a+b)
        print('Training loss :',cost)

        
        
    def Gradient_descent(self):
        for i in range(self.num_iters):
            temp= self.feed_forward()
            print('Iteration:',i)
            self.find_cost()
            self.W1= self.W1 - self.lr * self.DW1
            self.B1= self.B1 - self.lr * self.DB1
            self.W2= self.W2 - self.lr * self.DW2
            self.B2= self.B2 - self.lr * self.DB2
            clear_output(wait=True)
            
    def get_parameters(self):
        parameters=dict()
        parameters['w1']= self.W1
        parameters['w2']= self.W2
        parameters['b1']= self.B1
        parameters['b2']= self.B2
        return parameters
    
    def predict(self,parameters,X,val_y):

        W1= parameters['w1']
        B1= parameters['b1']
        W2= parameters['w2']
        B2= parameters['b2']
        
        Z1= np.dot(W1,X.T)+B1

        A1= np.tanh(Z1)

        Z2= np.dot(W2,A1)+B2
        
        A2= self.sigmoid(Z2)

        self.find_val_cost(val_y,A2)
        return A2
    def give_insight(self,val_y,answer):
        # print(confusion_matrix(val_y,answer))
        tn, fp, fn, tp= confusion_matrix(val_y,answer).ravel()
        # print('True Negative ', tn)
        # print('False Positive ', fp)
        # print('False Negative ', fn)
        # print('True Positive ', tp)

        total= tn+fp+fn+tn


        print('Accuracy: Overall, how often is the classifier correct?')
        print('Accuracy :', np.round((tp+tn)/total *100,2),"%\n")

        print('Misclassification Rate: Overall, how often is it wrong? ')
        print('Error rate',np.round((fp+fn)/total *100,2),'%\n')

        print('True Positive Rate: When its actually yes, how often does it predict yes? senstivity/recall')
        recall= (tp/(fn+tp) )*100
        print('True positive rate', np.round(recall),'%\n')

        print('False Positive Rate: When its actually no, how often does it predict yes? ')
        print('False positive rate',np.round(fp/ (tn+fp) *100,2),'%\n')

        print('True Negative Rate: When its actually no, how often does it predict no? ')
        print('True Negative rate',np.round(tn/ (tn+fp),2),'%\n')

        print('Precision: When it predicts yes, how often is it correct? ')
        precision= (tp/ (fp+tp)) *100
        print('precision',np.round( precision,2),'%\n')

        print('When it predicts no, how often is it correct? ')
        precision= (tn/ (fn+tn)) *100
        print(np.round( precision,2),'%\n')


        print('Prevalence: How often does the yes condition actually occur in our sample? ')
        print('prevalence',np.round((fn+tp)/total *100,2),'%\n')

        print('Prevalence: How often does the No condition actually occur in our sample? ')
        print('prevalence',np.round((tn+fp)/total *100,2),'%\n')


        print('F1 score: weighted average of the true positive rate (recall) and precision .  2* [ (precision * recall)/(precision+recall) ]')
        f1= 2* ((precision*recall)/(precision+recall))
        print('F1 score is:',np.round(f1,2),'% \n')
        return tn, fp, fn, tp
    
    def fit(self,X,Y,nodes):
        self.initialize_parameters(X,Y,nodes)
        self.Gradient_descent()
        self.find_cost()
        return self.get_parameters()

    def find_val_cost(self,Y,A2):
        N= A2.shape[1]
        a= np.ravel(np.dot(Y , np.log(A2).T))
        b= np.ravel(np.dot((1-Y),np.log(1-A2.T)))
        cost = -(1/N) * (a+b)
        print('validation loss :',cost)


tn=[]
fp=[]
fn=[]
tp=[]
itr=[]
lr=[]

# for num_iters in [500,1000,2000,3000,4000,5000]:
#     for learning_rate in np.arange(1,5,step=0.5):
#         hidden_nodes= 4
#         nn= NeuralNetwork_2layer_1H(num_iters,learning_rate)
#         final_parameters=nn.fit(train_x,np.array([train_y]),hidden_nodes)
#         answer= nn.predict(final_parameters,val_x,val_y)
#         tn, fp, fn, tp= nn.give_insight(val_y,np.squeeze((answer>0.5).tolist()).tolist())
#         tn.append(tn)
#         fp.append(fp)
#         fn.append(fn)
#         tp.append(tp)
#         lr.append(learning_rate)
#         itr.append(num_iters)

# pd.DataFrame({'iteration':itr,'tn':tn,'fp':fp,'fn':fn,'tp':tp})
#df.to_csv('model_log.csv')

num_iters=5000
learning_rate= 5
hidden_nodes= 4
nn= NeuralNetwork_2layer_1H(num_iters,learning_rate)
final_parameters=nn.fit(train_x,np.array([train_y]),hidden_nodes)
answer= nn.predict(final_parameters,val_x,val_y)
tn, fp, fn, tp= nn.give_insight(val_y,np.squeeze((answer>0.5).tolist()).tolist())
```
