## What role does length of tweet plays in Disaster tweet and Non-Disaster tweets ?

```python
#Comparing length of real tweet and non real tweetss
plt.style.use('fivethirtyeight')
train_group=train_set_orig.groupby('target')

real_text_length= [len(text) for text in train_group.get_group(1)['text']]
neg_text_length= [len(text) for text in train_group.get_group(0)['text']]

bins=120

plt.figure(figsize=(20,5))

plt.hist(neg_text_length,bins=bins,label='Negative',color='red',alpha=0.6)
plt.hist(real_text_length,bins= bins,label='Positive',color='green')

plt.ylabel('Number of characters')
plt.xlabel('Tweet count')
plt.legend()
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(10,5))
plt.title('Number of characters in real and Not Real tweets')
ax1.hist(real_text_length,color= 'green')
ax1.set_xlabel('Number of real tweets',color= 'green')
ax1.set_ylim([0,1000])
ax2.hist(neg_text_length,color='red')
ax2.set_xlabel('Number of not real tweets',color='red')
ax2.set_ylim([0,1000])
```

## Answer:
## People Tend to write tweets little long when they are talking about a Real Disaster
It can be clearly scene that tweets size between 120 to 400 length are same in both type of tweets.


```python
def avg(x):
    return np.sum([len(word) for word in x])

neg_tweet_avg=  [np.sum(list(map(avg,text)))/len(text.split()) for text in train_group.get_group(0)['text']]
post_tweet_avg= [np.sum(list(map(avg,text)))/len(text.split()) for text in train_group.get_group(1)['text']]
```

```python
fig,(ax1,ax2)= plt.subplots(1,2,figsize=(15,5))
sns.distplot(neg_tweet_avg,color='red',ax=ax1)
ax1.set_title('negative tweets')
ax1.set_xlabel('average tweet length')
sns.distplot(post_tweet_avg,color='green',ax=ax2)
ax2.set_title('positive tweets')
ax2.set_xlabel('average tweet length')
```

```python
def layer_size(X,Y):
    n_x= X.shape[1]
    n_h= 4
    n_y= Y.shape[0]
    return n_x,n_h,n_y
n_x,n_h,n_y=layer_size(X,Y)
print(n_x)
print(n_h)
print(n_y)
```

```python
def initialize_parameter(n_x,n_h,n_y):
    w1= np.random.randn(n_h,n_x)*0.01
    b1= np.zeros((n_h,1))
    w2= np.random.randn(n_y,n_h)*0.01
    b2= np.zeros((n_y,1))
    
    parameters={'w1':w1,'b1':b1,'w2':w2,'b2':b2}
    return parameters
    
parameters=initialize_parameter(n_x,n_h,n_y)
print('w1 :',parameters['w1'].shape)
print('b1 :',parameters['b1'].shape)
print('w2 :',parameters['w2'].shape)
print('b2 :',parameters['b2'].shape)
```

```python
def sigmoid(z):
    return 1/ (1+ np.exp(-z))

def forward_propogation(X,parameters): # X(10, 5242)
    w1= parameters['w1'] #(4, 5242)
    b1= parameters['b1'] #(4, 1)
    w2= parameters['w2'] #(1, 4)
    b2= parameters['b2'] #(1, 1)

    z1= np.dot(w1,X.T)+b1
    a1= np.tanh(z1) #(4, 10)
    
    z2= np.dot(w2,a1)+b2 #(1,10)

    a2= sigmoid(z2) #(1,10)

    cache= {'a1':a1,'a2':a2,'z1':z1,'z2':z2}
    return a2,cache

a2,cache= forward_propogation(X,parameters)
# print('z1 :',cache['z1'].shape)
```

```python
def compute_cost(a2,Y,parameters): #a2 (1,10) #Y(1, 10)
    m= Y.shape[1]

    cost= -(1/m) * np.sum( np.dot(Y, np.log(a2).T ) + np.dot((1-Y),np.log(1-a2).T) )
    return cost

cost= compute_cost(a2,Y,parameters)
```

```python
def backward_propogation(parameters,cache,X,Y): #X (10, 5242) #Y(1, 10)
    w1= parameters['w1'] #(4, 5242)
    b1= parameters['b1'] #(4, 1)
    w2= parameters['w2'] #(1, 4)
    b2= parameters['b2'] #(1, 1)
    
    z1= cache['z1'] #(4, 10)
    a1= cache['a1'] #(4, 10)
    
    z2= cache['z2'] #(1,10)
    a2= cache['a2'] #(1,10)
    
    m= Y.shape[1]
    
    dz2= a2 - Y #(1, 10)
    da2= sigmoid(dz2)
    dw2= np.dot(dz2,a1.T) * 1/m #(1, 4)
    db2= np.sum(dz2,axis=1,keepdims=True) * (1/m) #(1,1)
    
    dz1= np.dot(w2.T,dz2)* (1- a1**2) * (1/m) #(4, 10)
    dw1= np.dot(dz1,X) * (1/m)
    db1= np.sum(dz1,axis=1,keepdims=True) * (1/m)    
    grads= {'dw1':dw1,'dw2':dw2,'db1':db1,'db2':db2}
    return grads

    
    
grads=backward_propogation(parameters,cache,X,Y)
# backward_propogation(parameters,cache,X,Y)
```

```python
def update_parameters(parameters,grads,alpha):
    w1= parameters['w1']
    w2= parameters['w2']
    b1= parameters['b1']
    b2= parameters['b2']
    
    dw1= grads['dw1']
    dw2= grads['dw2']
    db1= grads['db1']
    db2= grads['db2']
    
    w1= w1 - alpha * dw1
    w2= w2 - alpha * dw2
    b1= b1 - alpha * db1
    b2= b2 - alpha * db2
    
    parameters={'w1':w1,'w2':w2,'b1':b1,'b2':b2}
    return parameters
parameters= update_parameters(parameters,grads,0.9)
```

```python
def nn_model(X,Y,n_h, num_iterations,alpha):
    n_x,n_h,n_y= layer_size(X,Y)
    parameters= initialize_parameter(n_x,n_h,n_y)
       
    for i in range(num_iterations):
        a2,cache= forward_propogation(X,parameters)
        cost = compute_cost(a2,Y,parameters)
        grads= backward_propogation(parameters,cache,X,Y)
        parameters= update_parameters(parameters,grads,alpha)
        if i%10 == 0:
            print(f'iteration {i} --> {cost} cost')
    return parameters

best_parameters= nn_model(X,Y,4,10000,6)
print('Model Successfully ran !!')
```

```python
def predict(parameters,test):
    a2,cache = forward_propogation(test,parameters)
    answer= a2>0.5
    return answer

y=predict(best_parameters,test)
```
